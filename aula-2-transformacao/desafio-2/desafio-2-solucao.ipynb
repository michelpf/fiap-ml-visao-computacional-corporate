## Desafio 2


Dado a imagem abaixo (no repositório “imagens/gize.png”), realize a extração da pirâmide, por meio de uma máscara poligonal, para uma outra imagem com fundo na cor verde. 

![alt text](imagens/gize.png)

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
print(cv2.__version__)

#Pop-up do Matplotlib
#%matplotlib qt

#Exibição na mesma tela do Jupyter
%matplotlib inline

plt.style.use('seaborn')
sns.set_style("whitegrid", {'axes.grid' : False})
```

```python
imagem = cv2.imread("imagens/gize.png")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(imagem)
plt.title("Pirâmide")
```

```python
imagem_colorida = imagem.copy()

height = imagem.shape[0]
width = imagem.shape[1]

mascara = np.zeros((height, width), dtype=np.uint8)
pontos = np.array([[60,410], [300,150], [520,420]], np.int32)
cv2.fillPoly(mascara, [pontos], (100, 50, 200))

plt.figure(figsize=(20,10))
plt.imshow(mascara, cmap="gray")
plt.title("Máscara de Pirâmide")
```

```python
imagem_branco = np.ones([height, width, 3])

plt.figure(figsize=(20,10))
plt.imshow(imagem_branco)
plt.title("Imagem Vazia")
```

*Neste [exemplo](https://docs.opencv.org/3.2.0/d0/d86/tutorial_py_image_arithmetics.html) da documentação do OpenCV detalha um pouco sobre o uso das máscaras.*

```python
# A máscara neste caso somente vai combinar na imagem passada os pixels diferente de 0, ou seja, 
# exatamente a pirâmide extraída

imagem_extraida = cv2.bitwise_and(imagem_colorida,imagem_colorida,mask=mascara)

plt.figure(figsize=(20,10))
plt.imshow(imagem_extraida)
plt.title("Imagem Extra'dia")
```

Substituindo o valor preto (zerado) para outra cor, neste caso verde.

```python
imagem_extraida[np.where((imagem_extraida == [0,0,0]).all(axis = 2))] = [0,255,0]

plt.figure(figsize=(20,10))
plt.imshow(imagem_extraida)
plt.title("Pirâmide Extraída")
```
